```python
from peptdeep.model.charge import *
import pandas as pd

model = ChargeModelForAASeq()

seq_df = pd.DataFrame({
    'sequence': ['ABCDE','FGHIJK','LMNOPQ','RSTUVWXYZ','HIJKL'],
    "charge" : [1, 2, 1, 2, 1],
})
seq_df = model.create_charge_indicators(seq_df,group_by_modseq=False)
model.train(seq_df)
model.predict_and_clip_charges(
    seq_df, 
    min_precursor_charge=1,
    max_precursor_charge=2,
    charge_prob_cutoff=0.3
)
```

```python
model.predict_charges_as_prob(seq_df, 2, 4)
```

```python
seq_df['charge'] = 1
model.predict_prob_for_charge(seq_df)
```

```python
from peptdeep.model.charge import *
import pandas as pd

model = ChargeModelForModAASeq()

modseq_df = pd.DataFrame({
    'sequence': ['ABCDE','FGHIJK','LMNOPQ','RSTUVWXYZ'],
    'mods': ['', '', 'Oxidation@M', 'Phospho@T'],
    'mod_sites': ['', '', '2', '3'],
    'charge' : [1, 2, 1, 2],
})
modseq_df = model.create_charge_indicators(modseq_df, group_by_modseq=True)
model.train(modseq_df, epoch=20)
model.predict_and_clip_charges(
    modseq_df, 
    min_precursor_charge=1,
    max_precursor_charge=2,
    charge_prob_cutoff=0.3
)
```

```python
model.predict_charges_as_prob(modseq_df, 2, 4)
```

```python
modseq_df['charge'] = 1
model.predict_prob_for_charge(modseq_df)
```
